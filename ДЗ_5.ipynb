```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data = {
    'pclass': [3, 1, 3, 1, 3, 2, 1, 3, 1, 3],
    'sex': ['male', 'female', 'female', 'female', 'male', 'male', 'female', 'female', 'male', 'male'],
    'age': [22.0, 38.0, 26.0, 35.0, 35.0, 27.0, 19.0, None, 26.0, 32.0],
    'sibsp': [1, 1, 0, 1, 0, 0, 0, 1, 0, 0],
    'parch': [0, 0, 0, 0, 0, 0, 0, 2, 0, 0],
    'fare': [7.2500, 71.2833, 7.9250, 53.1000, 8.0500, 13.0000, 30.0000, 23.4500, 30.0000, 7.7500],
    'embarked': ['S', 'C', 'S', 'S', 'S', 'S', 'S', 'S', 'C', 'Q']
}

df = pd.DataFrame(data)

print(df.info())
print(df.describe())
print(df.describe(include=['O']))
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['age'].dropna(), bins=30, kde=True)
plt.title('Розподіл віку пасажирів')
plt.xlabel('Вік')
plt.ylabel('Кількість')
plt.show()

plt.figure(figsize=(10, 6))
sns.histplot(df['fare'], bins=30, kde=True)
plt.title('Розподіл плати за квиток')
plt.xlabel('Плата за квиток')
plt.ylabel('Кількість')
plt.show()

fig, axes = plt.subplots(1, 2, figsize=(14, 6))
sns.countplot(ax=axes[0], x='sibsp', data=df)
axes[0].set_title('Розподіл кількості братів/сестер')
axes[0].set_xlabel('Кількість братів/сестер')
axes[0].set_ylabel('Кількість')

sns.countplot(ax=axes[1], x='parch', data=df)
axes[1].set_title('Розподіл кількості батьків/дітей')
axes[1].set_xlabel('Кількість батьків/дітей')
axes[1].set_ylabel('Кількість')

plt.show()
```

```python

plt.figure(figsize=(10, 6))
sns.countplot(x='pclass', data=df)
plt.title('Розподіл класів пасажирів')
plt.xlabel('Клас')
plt.ylabel('Кількість')
plt.show()

plt.figure(figsize=(10, 6))
sns.countplot(x='sex', data=df)
plt.title('Розподіл статі пасажирів')
plt.xlabel('Стать')
plt.ylabel('Кількість')
plt.show()

plt.figure(figsize=(10, 6))
sns.countplot(x='embarked', data=df)
plt.title('Розподіл порту посадки')
plt.xlabel('Порт посадки')
plt.ylabel('Кількість')
plt.show()
```

```python
fig, axes = plt.subplots(2, 2, figsize=(14, 12))

sns.boxplot(ax=axes[0, 0], x='age', data=df)
axes[0, 0].set_title('Викиди у віковій ознаці')

sns.boxplot(ax=axes[0, 1], x='sibsp', data=df)
axes[0, 1].set_title('Викиди у кількості братів/сестер')

sns.boxplot(ax=axes[1, 0], x='parch', data=df)
axes[1, 0].set_title('Викиди у кількості батьків/дітей')

sns.boxplot(ax=axes[1, 1], x='fare', data=df)
axes[1, 1].set_title('Викиди у платі за квиток')

plt.show()
```

Числові ознаки:

age: Розподіл віку пасажирів нормальний, однак є пропущені значення.
fare: Є значні викиди, які можуть вказувати на пасажирів першого класу з високою платою за квиток.
sibsp та parch: Більшість пасажирів подорожували без братів/сестер або батьків/дітей.
Категоріальні ознаки:

pclass: Більшість пасажирів подорожували третім класом.
sex: Є більша кількість чоловіків, ніж жінок.
embarked: Більшість пасажирів сідали на корабель у порту S.
Викиди:

Основні викиди виявлені в ознаках fare та age
